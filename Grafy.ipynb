```python
import matplotlib.pyplot as plt
import numpy as np

# Dta z novch vpotov
labels = ['Human [0]', 'AI [1]']
precision = [0.8471, 0.8948]
recall = [0.9014, 0.8375]
f1_score = [0.8734, 0.8652]

# Pozcia stpcov
x = np.arange(len(labels))
width = 0.25

# Farby - odtiene modrej
colors = ['#6baed6', '#3182bd', '#08519c']

# Vykreslenie grafu
fig, ax = plt.subplots(figsize=(9, 6))
bars1 = ax.bar(x - width, precision, width, label='Precision', color=colors[0])
bars2 = ax.bar(x, recall, width, label='Recall', color=colors[1])
bars3 = ax.bar(x + width, f1_score, width, label='F1-score', color=colors[2])

# Nastavenia
ax.set_ylabel('Hodnota skre', fontweight='bold')
ax.set_title('Metriky DistilBERT', fontweight='bold', pad=20)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontweight='bold')
ax.set_ylim(0.80, 0.95)  # Upraven rozsah pre nov hodnoty
ax.set_yticks(np.arange(0.80, 0.96, 0.02))  # Jemnejia kla
ax.legend(loc='upper right', bbox_to_anchor=(1, 1))  # Presunut legenda

# Hodnoty nad stpcami (formtovan na 4 desatinn miesta)
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.4f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontweight='bold', fontsize=9)

# Vylepen mrieka
plt.grid(axis='y', linestyle=':', alpha=0.7)
plt.tight_layout()

# Uloenie grafu (voliten)
# plt.savefig('metriky_human_ai.png', dpi=300)

plt.show()
```
